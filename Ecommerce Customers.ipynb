```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y = df['Yearly Amount Spent']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd

# Assume X is your features and y is your target variable
model = LinearRegression()
model.fit(x, y)

# Calculate the residuals (difference between actual and predicted)
predictions = model.predict(x)
residuals = np.abs(y - predictions)

# Define a threshold for identifying outliers in residuals (e.g., top 5% of residuals)
threshold = np.percentile(residuals, 95)

# Filter out rows where residuals are greater than the threshold
df = df[residuals < threshold]
```
